```python
dataset = pd.read_csv("Mall_customers.csv")
```

# Mean of the Age from the Data set

```python
dataset.head(10)
```

```python
dataset["Age"].mean()
```

```python
mn = np.mean(dataset["Age"])
```

```python
sns.histplot(x="Age", data=dataset, bins=[i for i in range(0,71, 10)])
plt.plot([ mn for i in range(0, 100)],[i for i in range(0, 100)],c="red")
plt.axvline(x=mn, color='red', linestyle='--', label=f'Mean: {mn:.2f}')

plt.legend() # Show the label
plt.show()
```

```python
dataset["Annual Income (k$)"].median()
```

```python
md = np.median(dataset["Annual Income (k$)"])
md
```

```python
sns.histplot(x="Annual Income (k$)", data=dataset, bins=[i for i in range(0,71, 10)])
plt.plot([ md for i in range(0, 100)],[i for i in range(0, 100)],c="red")
plt.axvline(x=mn, color='red', linestyle='--', label=f'Mean: {mn:.2f}')
plt.axvline(x=md, color='blue', linestyle='--', label=f'Median: {md:.2f}')
plt.legend() # Show the label
plt.show()
```
