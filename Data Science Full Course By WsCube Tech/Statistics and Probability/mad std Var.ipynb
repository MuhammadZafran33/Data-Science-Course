```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sec_a = np.array([23,45, 52, 73, 92, 61])
sec_b = np.array([11,32, 59, 79, 98, 189])
no = np.array([1, 2, 3, 4, 5, 6])
```

```python
mean = np.mean(sec_a)
mean
```

```python
mean = np.mean(sec_b)
mean
```

```python
plt.figure(figsize=(8, 4))
plt.scatter(sec_a,no, label="Sec A")
plt.scatter(sec_b,no, color="red",  label="Sec B")
plt.plot(range(1,7), no, color="blue")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.scatter(sec_a, no, label="Sec A")
plt.scatter(sec_b, no, color="red", label="Sec B")

# Calculate mean of sec_a
mean_sec_a = [np.mean(sec_a)] * len(no)  # Create array of mean values with same length as 'no'
mean_sec_b = [np.mean(sec_b)] * len(no)
# Plot the mean line
plt.plot(mean_sec_a, no, color="blue", label="Mean of Sec A")
plt.plot(mean_sec_b, no, color="red", label="Mean of Sec B")
plt.legend()
plt.show()
```

# Mean Absolute Deviation

Definition: 

The average distance of each value from the mean. It measures how spread out the data is.

Formula:
MAD=n∑i=1n​∣xi​−xˉ∣​

```python
np.abs(sec_a - mean)
```

```python
np.sum(np.abs(sec_a - mean))
```

```python
mad_a = np.sum(np.abs(sec_a - mean))/len(sec_a)
mad_b = np.sum(np.abs(sec_b - mean))/len(sec_b)
```

```python
mad_a, mad_b
```

# Standard Deviation(std) and Varience

Definition:

The square root of variance. It measures how spread out values are from the mean (same units as original data).




```python
np.std(sec_a), np.std(sec_b)
```

```python
np.var(sec_a), np.var(sec_b)
```
