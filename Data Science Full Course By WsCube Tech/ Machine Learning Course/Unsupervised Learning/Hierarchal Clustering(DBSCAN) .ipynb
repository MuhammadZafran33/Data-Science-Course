```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
```

```python
x,y = make_moons(n_samples=1000,noise=0.5)
```

```python
x
```

```python
y
```

```python
df = {"data1":x[:,0], "data2":x[:,1], "output":y}
```

```python
dataset = pd.DataFrame(df)
```

```python
dataset.head(3)
```

```python
sns.scatterplot(x="data1",y="data2",data=dataset,hue="output",)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.2,min_samples=5)
db.fit_predict(dataset)
```

```python
db = DBSCAN(eps=0.2,min_samples=5)
dataset["Predict"] = db.fit_predict(dataset)
```

```python
sns.scatterplot(x="data1",y="data2",data=dataset,hue="Predict")
plt.show()
```

```python
sns.scatterplot(x="data1",y="data2",data=dataset,hue=y)
plt.show()
```
