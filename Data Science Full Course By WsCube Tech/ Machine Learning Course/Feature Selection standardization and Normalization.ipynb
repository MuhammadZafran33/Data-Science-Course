# Standardization Using StandardScaler
## What is Standardization?
### Convert features to have mean = 0 and standard deviation = 1.
## Formula: (X - mean) / std

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

#### fit(): Learn mean and std from data
#### transform(): Apply formula to scale data
#### Result: Features centered around 0 with same spread

```python
ss.fit(dataset[["CHits"]])
```

```python
ss.transform(dataset[["CHits"]])
```

```python
dataset["CHits_ss"] = pd.DataFrame(ss.transform(dataset[["CHits"]]), columns=["x"])
```

```python
dataset.head(5)
```

```python
dataset.describe()
```

 ## Before & After Standardization Comparison

#### Left plot: Original data (wider range)
#### Right plot: Standardized data (centered at 0, narrower range)
#### Why? Visualize the effect of standardization

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["CHits"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["CHits_ss"])

plt.show()
```

##### Same comparison using histograms with KDE (density curve)
##### Shows distribution shape remains the same, only scale changes
##### Purpose: Different visualization style for better understanding

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.title("Before")
sns.histplot(dataset["CHits"], kde=True)

plt.subplot(1,2,2)
plt.title("After")
sns.histplot(dataset["CHits_ss"], kde=True)

plt.show()
```

# Normalization (Min-Max scaler) Technique.........

## What is Normalization?
#### Convert features to range [0, 1].
#### Formula: (X - min) / (max - min)

```python
dataset.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms = MinMaxScaler()
```

```python
ms.fit(dataset[["CHits"]])
```

```python
ms.transform(dataset[["CHits"]])
```

```python
dataset["CHits_ms"]=ms.transform(dataset[["CHits"]])
dataset.head(5)
```

## Before & After Normalization Comparison

```python
plt.figure(figsize=(10,7))

plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["CHits"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["CHits_ms"])

plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.title("Before")
sns.histplot(dataset["CHits"], kde=True)

plt.subplot(1,2,2)
plt.title("After")
sns.histplot(dataset["CHits_ms"], kde=True)

plt.show()
```

# Quick Comparison Table
## Feature -------------------Standardization ----------------- Normalization 
#### Range-------------------------------  -∞ to +∞-------------------------------------------------0 to 1
#### Mean------------------------------------0---------------------------------------------------Not fixed
#### Std Dev-------------------------------1-----------------------------------------------------Not fixed
#### Formula----------------------------(X - mean) / std-------------------------------------(X - min) / (max - min)
#### When to Use-------------------Machine Learning (Regression, SVM, KNN)------------------------Neural Networks, Image Processing
#### Sensitive to----------------------------Outliers (affected by std)----------------------------Outliers (affected by min/max)

# Key Takeaways
#### ✅ Standardization - Centers data around 0, useful for algorithms using distance/variance
#### ✅ Normalization - Scales to [0,1], useful when features need bounded ranges
#### ✅ Why? - Different algorithms perform better with differently scaled data
#### ✅ Always visualize - Plots help understand the transformation effect
#### ✅ Fit on training, transform on test - Prevent data leakage in real projects


# When to Use What?
## Use Standardization When:

#### Using Linear Regression, SVM, or KNN
#### Data is normally distributed
#### Want to preserve outlier information

## Use Normalization When:

#### Using Neural Networks
#### Algorithm has bounded range requirement
#### Working with image/pixel data
#### Comparing features with different units
