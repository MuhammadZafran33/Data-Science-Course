```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv("car data.csv")
dataset.head(5)
```

```python
sns.scatterplot(x="Year", y="Selling_Price", data=dataset)
plt.show()
```

```python
sns.scatterplot(x="Year", y="Kms_Driven", data=dataset)
plt.show()
```

```python
dataset.isnull().sum()
```

```python
x = dataset[["Year"]]
x.ndim
```

```python
x = dataset[["Year"]]
y = dataset[["Selling_Price"]]
```

```python
x_train, x_test,y_train, y_test =  train_test_split(x,y, test_size=0.2, random_state=50)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.predict([[60]])
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
# Make predictions
y_pred = lr.predict(x)
```

```python


# Create the plot
plt.figure(figsize=(5,4))
sns.scatterplot(x="Year", y="Selling_Price", data=dataset)
plt.plot(dataset["Year"], y_pred, c="red")
plt.legend(["Original", "Prediction line"])
plt.show()
plt.savefig("LR Predicted line.jpg")
```

```python


# Create the plot
plt.figure(figsize=(5,4))
sns.scatterplot(x="Year", y="Kms_Driven", data=dataset)
plt.plot(dataset["Year"], y_pred, c="red")
plt.legend(["Original", "Prediction line"])
plt.show()
plt.savefig("LR Predicted line.jpg")
```
