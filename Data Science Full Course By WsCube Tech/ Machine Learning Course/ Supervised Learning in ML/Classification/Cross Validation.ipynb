```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/student_performance2000.csv')
```

```python
dataset.head(5)
```

```python
x =  dataset.iloc[:,:-1]
y = dataset["final_score"]
```

```python
new_data = dataset.head(10)
```

```python
from sklearn.model_selection import LeaveOneOut, LeavePOut, KFold, StratifiedKFold
```

```python
 lo = LeaveOneOut()
 for train, test in lo.split(x,y):
   print(train, test)
```

```python
x_new = new_data.iloc[:,:-1]
y_new = new_data["final_score"]
```

```python
 lo = LeaveOneOut()
 for train, test in lo.split(x_new, y_new):
   print(train, test)
```

```python
 lo = LeavePOut(p=3)
 for train, test in lo.split(x_new, y_new):
   print(train, test)
```

```python
 lo = KFold(n_splits=5)
 for train, test in lo.split(x_new, y_new):
   print(train, test)
```
