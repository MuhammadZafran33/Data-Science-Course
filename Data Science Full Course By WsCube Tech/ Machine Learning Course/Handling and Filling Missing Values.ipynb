# Filling Null values in a dataset


```python
dataset = pd.read_csv("anime.csv")
```

```python
dataset.isnull().sum()
```

```python
dataset.head(10)
```

### This is a WRONG Method to use because no datatype Declear

```python
dataset.fillna(10)
```

```python
dataset.isnull().sum()
```

## Best Methods to Fill Null values in dataset are.............

## Backfill Method....

```python
dataset.fillna(method='bfill')
```

## Farword Methode.....

```python
dataset.fillna(method='ffill')
```

## Farword Filling using Row wise...

```python
dataset.fillna(method='ffill', axis=1)
```

## Backward Filling using Row wise...

```python
dataset.fillna(method='bfill', axis=1)
```

```python
dataset.isnull().sum()
```

```python
dataset.info()
```

# Mode Method Filling Missing values.............

```python
dataset["rating"].mode()[0]
```

```python
dataset["rating"].fillna(dataset["rating"].mode()[0], inplace=True)
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

## Using for loop to fill all the missing by mode of that Column Exactly.........

```python
for i in dataset.select_dtypes(include="object").columns:
    dataset[i].fillna(dataset[i].mode()[0], inplace=True)
```

```python
dataset.isnull().sum()
```
