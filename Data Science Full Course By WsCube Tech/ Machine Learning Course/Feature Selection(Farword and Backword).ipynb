```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.linear_model import LinearRegression
import numpy as np

# Dummy data
X = np.random.random((100, 5))
y = np.random.random(100)

# We add scoring='r2' to avoid the attribute lookup error
sfs = SFS(LinearRegression(), 
          k_features=3, 
          forward=True, 
          floating=False, 
          scoring='r2',  # <--- Add this!
          cv=2)

sfs.fit(X, y)

print("Selected indices:", sfs.k_feature_idx_)
```

```python
dataset = pd.read_csv("Anime.csv")
```

```python
dataset.head(3)
```

```python
# Assuming you already have x and y defined from dataset
x = dataset.iloc[:,:-1]
y = dataset["members"]

print(x.shape)  # Should output: (768, 5)
```

```python
from sklearn.linear_model import LogisticRegression
from mlxtend.feature_selection import SequentialFeatureSelector




```

```python
# Create logistic regression model
lr = LogisticRegression()
```

```python
# Create SequentialFeatureSelector
fs = SequentialFeatureSelector(lr, k_features=5, forward=True)
```

```python
# Fit the feature selector
fs.fit(x, y)
```

```python
# To get the selected features
selected_features = fs.k_feature_idx_
print("Selected feature indices:", selected_features)
```
