```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Student_performance.csv')
```

```python
dataset.head(4)
```

```python
x = dataset.iloc[:,:-1]
y = dataset["final_score"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_train,y_train)*100, lr.score(x_test,y_test)*100
```

```python
dt = LinearRegression()
dt.fit(x_train,y_train)
dt.score(x_train,y_train)*100, dt.score(x_test,y_test)*100
```

```python
sv = SVR()
sv.fit(x_train,y_train)
sv.score(x_train,y_train)*100, sv.score(x_test,y_test)*10
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
li = [("lr1", LinearRegression()), ("dt1", DecisionTreeRegressor()), ("sv1", SVR())]
vc = VotingRegressor(li)
vc.fit(x_train,y_train)

```

```python
vc.score(x_train,y_train)*100, vc.score(x_test,y_test)*100
```

```python
df = {"lr": lr.predict(x_test), "dt": dt.predict(x_test), "sv": sv.predict(x_test), "vc": vc.predict(x_test)}
```
