```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/diabetes_dataset.csv")
```

```python
dataset.head(4)
```

```python
sns.scatterplot(x="BloodPressure", y="Insulin", data=dataset, hue="Outcome")
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["Outcome"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import Perceptron
```

```python
pr = Perceptron(alpha=0.01)
pr.fit(x_train, y_train)
```

```python
pr.score(x_train, y_train)*100, pr.score(x_test, y_test)*100
```

```python
plt.figure(figsize=(10, 8))
filler_values = {x_train.columns.get_loc(col): x_train[col].mean() for col in x_train.columns if col not in ['BloodPressure', 'Insulin']}
plot_decision_regions(X=x_train.values, y=y_train.values, clf=pr, feature_index=[4,5], filler_feature_values=filler_values, legend=2, scatter_kwargs={'s': 80, 'alpha': 0.9, 'edgecolor': 'black'})
plt.xlabel('BloodPressure')
plt.ylabel('Insulin')
plt.title('Perceptron Decision Regions with Data Points')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(x=x_train['BloodPressure'], y=x_train['Insulin'], hue=y_train, s=80, alpha=0.9, edgecolor='black')
plt.title('Diagnostic Scatter Plot of BloodPressure vs Insulin')
plt.xlabel('BloodPressure')
plt.ylabel('Insulin')
plt.grid(True)
plt.tight_layout()
plt.show()
```
